## 2. JoinPlot
jointplot is a function used to create a multi-panel figure that displays both a univariate distribution of each variable along the diagonal axes and a bivariate visualization of the relationship between two variables in the off-diagonal panels.

**Univariate Distributions:**

Along the diagonal of the plot, separate univariate histograms or kernel density estimates are plotted for each variable. These histograms or kernel density estimates show the distribution of each variable individuall

**Bivariate Relationship:**

In the off-diagonal panels, a bivariate scatter plot or hexbin plot is displayed to show the relationship between the two variables. This allows you to visualize how the variables interact with each other.y.

```python
sns.jointplot(x="tip", y="total_bill", data = data, kind = "hex")
```

```python
# here type is regression
sns.jointplot(x="tip", y="total_bill", data = data, kind = "hex")
```
