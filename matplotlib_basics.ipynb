### Coding Style - Two ways to implement Matplotlib


#### 1. Explicit - Object oriented style


```python
x = np.linspace(0,2,100)
fig, ax = plt.subplots(figsize=(5,2.7), layout='constrained')
ax.plot(x,x, label="Linear")
ax.plot(x,x**2, label="Quadratic")
ax.plot(x,x**3, label ="Qubic")
ax.set_xlabel("X Label")
ax.set_ylabel("Y Label")
ax.set_title("Sample plot")
ax.legend()
```

#### 2. Implicit - pyplot function

```python
x= np.linspace(0,2,100)
plt.plot(x,x, label = "Linear")
plt.plot(x,x**2, label = "Quadratic")
plt.plot(x,x**3, label = "Qubic")
plt.xlabel("X Label")
plt.ylabel("Y Label")
plt.title("Simple plot")
plt.legend()
```

### bar graphs
With Pyplot, you can use the
**bar()**
function to draw bar graphs:



```python
x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])
plt.bar(x,y)
```

#### Horizontal BarsIf you want the bars to be displayed horizontally instead of vertically, use the barh() function

```python
x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])
plt.barh(x,y)
```

#### Bar Color
The bar() and barh() take the keyword argument color to set the color of the bars:

```python
x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])
plt.bar(x,y,color="red")
```

#### Bar Width

```python
x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])
plt.bar(x,y,width=.1)
```

##### Note: For horizontal bars, use height instead of width. #####


```python
x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])
plt.barh(x,y,height=.1)
```

### Pie Charts
With Pyplot, you can use the pie() function to draw pie charts:



```python
y = np.array([35, 25, 25, 15])
plt.pie(y)
```

##### As you can see the pie chart draws one piece (called a wedge) for each value in the array (in this case [35, 25, 25, 15]). #####

##### By default the plotting of the first wedge starts from the x-axis and moves counterclockwise: #####

```python
y = np.array([35, 25, 25, 15])
my_labels = ["apple", "banana", "orange", "pinapple"]
plt.pie(y, labels=my_labels)
```

#### Start Angle

As mentioned the default start angle is at the x-axis, but you can change the start angle by specifying a startangle parameter

The startangle parameter is defined with an angle in degrees, default angle is:



```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
plt.pie(y, labels=mylabels, startangle=90)
```

#### Explode
**To stand out wedges, The explode parameter allows you to do that.**



```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
myexplode = [.2,0,0,0]
plt.pie(y, labels = mylabels, explode = myexplode )
```

#### Shadow
Add a shadow to the pie chart by setting the shadows parameter to True:



```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
myexplode = [.2, 0, 0, 0]
plt.pie(y,labels = mylabels, explode = myexplode, shadow = True)
```

#### Colors
You can set the color of each wedge with the colors parameter.

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
mycolors = ["black", "hotpink", "b", "#4CAF50"]
plt.pie(y,labels = mylabels, colors = mycolors)
```

#### Legend 
To add a list of explanation for each wedge, use the legend() function:

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
plt.pie(y, labels = mylabels)
plt.legend()
```

#### Legend With Header


```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
plt.pie(y, labels = mylabels)
plt.legend(title = "Fruits")
```
